```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)

y=z * np.sin(20 * z)
x= z * np.cos(20 * z)

ax.plot3D(x, y, z, 'darkorange')
ax.set_title('3D line plot')

plt.savefig('3D.png')
```
