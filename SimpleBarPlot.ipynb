```python
import matplotlib.pyplot as plt
%matplotlib inline
#creating a new figure
fig = plt.figure()
#adding axes to the figure
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP','R']
students = [5000,6300,10000,9000,3000,4000]
#Y-label
plt.ylabel('Students')
#X-lable
plt.xlabel('Programming languages')
#title
plt.title('No. of Students learning Programming language')
#creating bar-plot
ax.bar(langs,students,width=0.5,color='skyblue',alpha=0.7)
#saving the figure in .png format
plt.savefig('plot.png', dpi=300, bbox_inches='tight')
plt.savefig('simplebarplot.png')
```

```python

import numpy as np
import matplotlib.pyplot as plt

data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
label=['2019','2020','2021','2022']
X = np.arange(len(label))
y=[0.25,1.25,2.25,3.25]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Comparing sales of three years in different cities')
plt.xticks(y,label)
#creating bar plots
ax.bar(X + 0.00, data[0], color = 'black', width = 0.25,alpha=0.5,label='Bengaluru')
ax.bar(X + 0.25, data[1], color = 'orange', width = 0.25,alpha=0.6,label='Delhi')
ax.bar(X + 0.50, data[2], color = 'red', width = 0.25,alpha=0.5,label='Chennai')
plt.legend()
plt.show()
```

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
clothing = (20, 35, 30, 35, 27)
accessories = (25, 32, 34, 20, 25)
ind = np.arange(5)    # the x locations for the groups
plt.bar(ind, clothing, width= 0.35 ,color='black',alpha=0.8,label='clothing')
plt.bar(ind, accessories, width= 0.35 ,bottom=clothing,color='orange',label='accessories')
plt.xlabel('Store loction')
plt.ylabel('Revenue generated(in thousands)')
plt.title('Revenue generated in cities according to clothes and accessories')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend()
plt.show()
```
