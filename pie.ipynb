```python
import matplotlib.pyplot as plt
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'English', 'Spanish', 'Russian', 'French','Chinese'
sizes = [55, 15, 8, 10,12]
explode = (0, 0.3, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
color={'green', 'orange','red','c','pink'}
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, colors=color,labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90,radius=1)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('pie.png', dpi=30, bbox_inches='tight')
plt.show()
```
