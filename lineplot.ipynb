```python
import matplotlib.pyplot as plt
```

```python
sales1=[2,4,5,6,4,24,20,13,14,22]
sales2=[1,4,3,5,9.6,13,9,14,12,23]
```

```python
linegraph1=plt.plot(sales1,color='r',alpha=0.6)
linegraph2=plt.plot(sales2,color='blue',alpha=0.6)
point1=plt.plot(sales1,'C0o', alpha=0.5,color='black')
point2=plt.plot(sales2,'C0o', alpha=0.5,color='orange')
plt.title('Monthly sales of 2022 and 2023')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.legend(['year 2018', 'year 2019'], loc=4)
plt.savefig('lineplot1.png')
```
