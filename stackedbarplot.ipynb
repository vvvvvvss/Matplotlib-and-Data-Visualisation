```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
clothing = (20, 35, 30, 35, 27)
accessories = (25, 32, 34, 20, 25)
ind = np.arange(5)    # the x locations for the groups
plt.bar(ind, clothing, width= 0.35 ,color='black',alpha=0.8,label='clothing')
plt.bar(ind, accessories, width= 0.35 ,bottom=clothing,color='orange',label='accessories')
plt.xlabel('Store loction')
plt.ylabel('Revenue generated(in thousands)')
plt.title('Revenue generated in cities according to clothes and accessories')
plt.xticks(ind, ('Bengaluru', 'Delhi', 'Chennai', 'Hyderabad', 'Mumbai'))
plt.yticks(np.arange(0, 81, 10))
plt.legend()
plt.savefig('stacked.png')
```
